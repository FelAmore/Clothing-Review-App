### 1.2 Pre-processing data
Perform the required text pre-processing steps.

#### Tokenizing each Review

```python
def tokenize_review(review):
    # segament into sentences
    sentences = sent_tokenize(review)
    
    # tokenize each sentence
    pattern = r"[a-zA-Z]+(?:[-'][a-zA-Z]+)?"
    tokenizer = RegexpTokenizer(pattern) 
    token_lists = [tokenizer.tokenize(sen) for sen in sentences]
    
    # merge them into a list of tokens
    tokenised_review = list(chain.from_iterable(token_lists))
    return tokenised_review
```

```python
tk_reviews = [tokenize_review(review) for review in df['Review Text']]

emp = 10
# preview output
print("Original review:")
print(df['Review Text'][emp])
print("Tokenized review:")
print(tk_reviews[emp])
```

```python
# checking stats after tokenization
def print_stats(tk_reviews):
    words = list(chain.from_iterable(tk_reviews))
    vocab = set(words)
    lexical_diversity = len(vocab)/len(words)
    print("Vocabulary size: ", len(vocab))
    print("Total number of tokens: ", len(words))
    print("Lexical diversity: ", lexical_diversity)
    print("Total number of reviews:", len(tk_reviews))
    lens = [len(review) for review in tk_reviews]
    print("Average review length:", np.mean(lens))
    print("Maximum review length:", np.max(lens))
    print("Minimum review length:", np.min(lens))
    print("Standard deviation of review length:", np.std(lens))

print_stats(tk_reviews)
```

#### Converting to Lower Case

```python
tk_reviews_lowered = [[word.lower() for word in review] for review in tk_reviews]

# preview output
print("Before lowercased:")
print(tk_reviews[emp])
print("After lowercased:")
print(tk_reviews_lowered[emp])
```

```python
# checking stats after lowercasing
print_stats(tk_reviews)
print()
print_stats(tk_reviews_lowered)
# vocab size decreased 
```

#### Removing Single Character Word

```python
tk_reviews_rsw = [[word for word in review if len(word) > 1] for review in tk_reviews_lowered]

# preview output
print("Before removing:")
print(tk_reviews_lowered[emp])
print("After removing:")
print(tk_reviews_rsw[emp])
# the single character word like 'a' and 'i' are removed
```

```python
# checking stats after removing single character words
print_stats(tk_reviews_lowered)
print()
print_stats(tk_reviews_rsw)
# vocab size and number of tokens decreased
```

#### Removing Stop Words

```python
# load stopwords
with open('stopwords_en.txt', 'r') as f:
    stopwords = set(f.read().splitlines())

# previewing stop words
print(f"Length of stopwords: {len(stopwords)}")
print("Preview:", list(stopwords)[:10])
```

```python
# remove stopwords based on the list given
tk_reviews_rstop = [[word for word in review if word not in stopwords] for review in tk_reviews_rsw]

# preview output
print("Before removing:")
print(tk_reviews_rsw[emp])
print("After removing:")
print(tk_reviews_rstop[emp])
```

```python
# checking stats after removing stopwords
print_stats(tk_reviews_rsw)
print()
print_stats(tk_reviews_rstop)
# vocab size and number of tokens decreased
# lexical diversity increased
```

#### Removing Words that Only Appears Once

```python
# compute document frequency for each unique word/type
words = list(chain.from_iterable(tk_reviews_rstop))
term_fd = FreqDist(words)

# find words that appear only once 
lessFreqWords = set(term_fd.hapaxes())
print(f"Length of words appearing only once: {len(lessFreqWords)}")
print("Preview", lessFreqWords)
```

```python
# remove lessFreqWords
tk_reviews_rhapaxes = [[word for word in review if word not in lessFreqWords] for review in tk_reviews_rstop]

# preview output
print("Before removing:")
print(tk_reviews_rstop[emp])
print("After removing:")
print(tk_reviews_rhapaxes[emp])
# lessFreqWords like 'grrrrrrrrrrr' are removed
```

```python
# checking stats after removing hapaxes
print_stats(tk_reviews_rstop)
print()
print_stats(tk_reviews_rhapaxes)
# vocab size, number of tokens, and lexical diversity decreased
```

#### Removing Top 20 Most Frequent Words

```python
# compute document frequency for each unique word/type
words_df = list(chain.from_iterable([set(review) for review in tk_reviews_rhapaxes]))
doc_fd = FreqDist(words_df)

# find the top 20 most frequent words 
top20 = set([word for word, freq in doc_fd.most_common(20)])
print("Top 20:")
print(doc_fd.most_common(20))
```

```python
# remove the top 20 mostFreqWords
tk_reviews_top20 = [[word for word in review if word not in top20] for review in tk_reviews_rhapaxes]

# preview output
print("Before removing:")
print(tk_reviews_rhapaxes[emp])
print("After removing:")
print(tk_reviews_top20[emp])
# words like bought, dress, fit are removed
```

```python
# checking stats after removing top 20 mostFreqWords
print_stats(tk_reviews_rhapaxes)
print()
print_stats(tk_reviews_top20)
# vocab size and number of tokens decreased
# lexical diversity increased
```

#### Saving Processed Data

```python
# joining tokens into string separated by space
joined = [' '.join(review) for review in tk_reviews_top20]

# preview output
print("Length:", len(joined)) # equals to number of reviews
print("Preview:", joined[emp])
```

```python
# adding new column 
df['Processed Review Text'] = joined

# preview output
print(df.shape)
print(df.head())
```

```python
# saving to csv
df.to_csv('processed.csv', index=False)

# verifying saved content
df_check = pd.read_csv('processed.csv')
print(df_check.head())
```

## Saving required outputs
Save the requested information as per specification.
- vocab.txt

```python
# generating the vocabulary
words = list(chain.from_iterable(tk_reviews_top20)) # put all the tokens in the corpus in a single list
vocab = sorted(list(set(words)))  # compute vocabulary and sort it alphabetically

print(f"Vocab Length: {len(vocab)}") # equals to the vocab size of tk_reviews_top20
print("Preview:")
print(vocab[:10])
```

```python
# index starting from 0
vocab_dict = {word: idx for idx, word in enumerate(vocab)}

# preview output
for word, idx in list(vocab_dict.items())[:10]:
    print(f"{word}:{idx}")
```

```python
# save vocab in the required format
with open('vocab.txt', 'w') as f:
    for word, idx in vocab_dict.items():
        f.write(f"{word}:{idx}\n")

# verifying saved content
with open('vocab.txt', 'r') as f:
    lines = f.read().splitlines()

print()
print(f"Length of vocab.txt: {len(lines)}")
print("Preview:")
for line in lines[:10]:
    print(line)
```
